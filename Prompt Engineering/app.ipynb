```python
import google.generativeai as genai

# Add your API key
genai.configure(api_key="API_KEY")

# Load model
model = genai.GenerativeModel("gemini-2.5-flash")


# Function to analyze logs
def analyze_log(log):
    prompt = f"""
    You are an expert cybersecurity analyst.

    Task:
    1. Classify the log into one of these:
       - Normal
       - Suspicious
       - Malicious
    2. Give a short explanation (1-2 lines)

    Examples:

    Log: User logged in successfully
    Answer: Normal - This is a standard login activity.

    Log: 10 failed login attempts in 1 minute
    Answer: Suspicious - Possible brute-force attempt.

    Now analyze:

    Log: {log}
    Answer:
    """

    response = model.generate_content(prompt)
    return response.text


# Run program
if __name__ == "__main__":
    while True:
        user_input = input("Enter log (or 'exit'): ")
        if user_input.lower() == "exit":
            break

        result = analyze_log(user_input)
        print("\nResult:")
        print(result)
        print("-" * 50)
```
